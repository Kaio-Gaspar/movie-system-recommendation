```python

from sklearn.preprocessing import MultiLabelBinarizer
import ast

# Converter strings de dicionários para listas de dicionários
movie_df['genres'] = movie_df['genres'].apply(ast.literal_eval)

# Extrair os nomes dos gêneros
generos = movie_df['genres'].apply(lambda x: [genero['name'] for genero in x])

# Inicializar o codificador one-hot
mlb = MultiLabelBinarizer()

# Ajustar e transformar os dados
generos_encoded = mlb.fit_transform(generos)

# Substituir a coluna original pela codificação one-hot dos gêneros
movie_df.drop(columns=['genres'], inplace=True)  # Remover a coluna original
df_generos_encoded = pd.DataFrame(generos_encoded, columns=mlb.classes_)  # Criar um novo DataFrame com as características (features) codificadas
movie_df = pd.concat([movie_df, df_generos_encoded], axis=1)  # Concatenar o DataFrame original com o DataFrame das características (features) codificadas
```

```python
y = movie_df['title']
X = movie_df.drop(['title','id'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Padronizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Treinar o modelo KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Fazer previsões
y_pred = knn.predict(X_test_scaled)
y_test,
y_pred
```

```python


# Calcular a precisão
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
